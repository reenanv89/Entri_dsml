Exercise 4: Lambda Function to Check if value is in a List


```python
#Without Lambda function
L=[2,4,6,7,9,45]

def func(L):
   num=7
   if num in L:
      print("Element is present in the list")
   else:
      print("Element is not present in the list")

func(L)
```

```python
#with lambda function
lst = [2, 4, 6, 7, 9, 45]
num = 8

func = lambda lst: print("Element is present in the list") if num in lst else print("Element is not present in the list")

func(lst)
```

Exercise 6 :
Write a python function, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. Return the results as a list


```python
def even(x,y):
   result=[num for num in range(x,y+1) if all(int(digits)%2==0 for digits in str(num))]
   return result

result=even(1000,3000)
print(result)
```

```python
def sentence(x):
  alpha=0
  num=0
  for sent in x:
      if sent.isalpha():
          alpha+=1
      elif sent.isnumeric():
          num+=1
  return alpha ,num
alpha,num=sentence("Hello world!123")
print("LETTERS ",alpha)
print("DIGITS",num)
```

Using lambda and map function

```python
sentence = lambda x: (sum(map(str.isalpha, x)), sum(map(str.isnumeric, x)))

# Call the lambda function with a sample string
alpha, num = sentence("HAPPY NEW YEAR 2024")

# Print the counts
print("LETTERS:", alpha)
print("DIGITS:", num)
```

8.Write a Python program to convert all the characters into uppercase and lowercase and eliminate duplicate letters from a given sequence. Use the map() function


```python
# Convert all the characters into upper and lower case and eliminate duplicate letters in a sequence
def change(x):
    return str.upper(x),str.lower(x)

sequence={"a","b","r","a","c","a","d,","a","b","r","a"}
result=map(change,sequence)
print(set(result))
```

Exercise 9 MAP:
 Write a Python program to add two given lists and find the difference between them. Use the map() function


```python
l1=[2,4,6,7,9]
l2=[1,2,3,4,5,4]
def sum(a,b):
    return a+b
def diff(a,b):
    return a-b

sum_result=list(map(sum,l1,l2))
diff_result=list(map(diff,l1,l2))
print("Sum of Lists=",sum_result)
print("Difference of lists=",diff_result)
```

using lambda function

```python
l1=[2,4,6,7,9]
l2=[1,2,3,4,5,4]

sum_result=list(map(lambda x,y:x+y,l1,l2))
diff_result=list(map(lambda x,y:x-y,l1,l2))
print("Sum of Lists=",sum_result)
print("Difference of lists=",diff_result)
```

Exercise 14 Reduce:


Write a Python program to calculate the average value of the numbers in a given tuple of tuples using lambda.
Original Tuple:
((10, 10, 10), (30, 45, 56), (81, 80, 39), (1, 2, 3))
Average value of the numbers of the said tuple of tuples:
(30.5, 34.25, 27.0)


```python
Original_Tuple= ((10, 10, 10), (30, 45, 56), (81, 80, 39), (1, 2, 3))
y=tuple(zip(*Original_Tuple))
avg=tuple(map(lambda x:sum(x)/len(x),y))
print("Original Tuple:\n", Original_Tuple)
print("Average Value of numbers of the said tuple of tuples:\n",avg)
```

Exercise 16:
Write a Python program to count the occurrences of items in a given list using lambda.
Original list:
[3, 4, 5, 8, 0, 3, 8, 5, 0, 3, 1, 5, 2, 3, 4, 2]
Count the occurrences of the items in the said list:
{3: 4, 4: 2, 5: 3, 8: 2, 0: 2, 1: 1, 2: 2}






```python
Original_list=[3, 4, 5, 8, 0, 3, 8, 5, 0, 3, 1, 5, 2, 3, 4, 2]
result=dict(map(lambda x:(x, Original_list.count(x)),Original_list))
print("Original list=",Original_list)
print("numbers:counts=",result)
```
